```python
import numpy as np
def grad_descent(x,y):
    mc = bc = 0
    iteration = 5000
    n = len(x)
    lr = 0.001 #learning rate
    
    for i in range(iteration):
        ypred = mc * x + bc
        cost = (1/n) * sum([val**2 for val in (y-ypred)])
        md = -(2/n) * sum(x*(y-ypred))
        bd = -(2/n) * sum(y-ypred)
        mc = mc - lr * md
        bc = bc - lr * bd
        print("m {}, b {}, cost {}, iteration {} ".format(mc, bc, cost, i))
        
        
        
```

```python
import math
def gradient_descent(x,y):
    mc = bc = 0
    iteration = 1000  #needs to be increased
    n = len(x)
    lr = 0.0002 #learning rate thresshold value
    cost_previous = 0
    for i in range(iteration):
        ypred = mc * x + bc
        cost = (1/n) * sum([val**2 for val in (y-ypred)])
        md = -(2/n) * sum(x*(y-ypred))
        bd = -(2/n) * sum(y-ypred)
        mc = mc - lr * md
        bc = bc - lr * bd
        if math.isclose(cost, cost_previous, rel_tol=1e-20):
            break
        cost_previous = cost
        print("m {}, b {}, cost {}, iteration {} ".format(mc, bc, cost, i))     
```
